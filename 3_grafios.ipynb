```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# grafica x al cuadrado

x = np.linspace(-10, 10, 500)
y = x**2
plt.plot(x,y,"m", label="x^2")
plt.title("Funcion cuadratica", fontname="Arial", fontsize=18)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

```python


horas=[2,8,14,16,20]
puntaje = [20,28,56,64,78]
plt.scatter(horas,puntaje)
plt.title("horas de estudio vs puntaje en evalucaion")
plt.xlabel("horas de estudio")
plt.ylabel("puntaje (0-100)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
materias = ["Matemáticas", "Física", "Química", "Inglés"]
notas = [85, 55, 90, 40]

plt.bar(materias, notas)
plt.title("Notas por materia")
plt.xlabel("Materias")
plt.ylabel("Nota (0-100)")
plt.ylim(0,100)
plt.grid(axis="y")
plt.show()
```

```python
porcentajes = [40, 25, 20, 15]
actividades = ["Estudio", "Deporte", "Redes Sociales", "Descanso"]

plt.pie(porcentajes, labels=actividades, autopct="%1.1f%%")
plt.title("Distribución del tiempo diario")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 500)
y = np.sin(x)

plt.plot(x, y, "r")
plt.title("Comportamiento del Movimiento")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```
